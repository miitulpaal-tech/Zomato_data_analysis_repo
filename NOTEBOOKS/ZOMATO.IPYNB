```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

- 1.LOAD THE DATASET

```python

data = pd.read_csv('zomato.csv')
```

```python
data.info()
```

- 2.REMOVING REDUNDANT COLUMNS

```python
data_2 = data.drop(columns =['url','address','phone','listed_in(city)','dish_liked','menu_item','location'])
```

- 3.Rename to sensible column_names


```python
data_2 = data_2.rename(columns={
    'rate': 'rating',
    'rest_type': 'restaurant_type',
    'approx_cost(for two people)': 'cost_for_2',
    'listed_in(type)': 'restaurant_type_category'
})
```

- 4.Drop null values from particular columns

```python
data_2 = data_2.dropna(subset = ['cuisines','cost_for_2'])
```

- 5.drop duplicate values

```python
data_2 = data_2.drop_duplicates()
```

- 6.Changing data type of string columns

```python
data_2['cost_for_2']  = data_2['cost_for_2'].str.replace(',','').astype(int)
```

- AIRTHEMATIC CHANGES AND CREATION OF NEW COLUMN

```python
data_2['cost_for_one'] = data_2['cost_for_2']//2
```

- RENAMING COLUMNS

```python
data_2[['name',
      'online_order',
      'book_table',
      'rating',
      'restaurant_type',
      'cuisines',
      'reviews_list',
      'restaurant_type_category']] = (
    data_2[['name',
          'online_order',
          'book_table',
          'rating',
          'restaurant_type',
          'cuisines',
          'reviews_list',
          'restaurant_type_category']].astype('string')
)
```

```python
data_3 = data_2
```

```python
data_3.drop(columns = ['cost_for_2'])
```

- function to handle rating column

```python
def handlerate(value):
  if(value == 'NEW' or value == '-' or value == "NA" ):
    return np.nan
  else:
    value = str(value).split('/')
    value = value[0]
    return float(value)
```

```python
data_3['rating'] = data_3['rating'].str.split('/').str[0].str.strip()
data_3['rating'] = pd.to_numeric(data_3['rating'], errors='coerce')
data_3['rating'] = data_3['rating'].fillna(data_3['rating'].mean())
```

- Applying the function

```python
data_3['rating'] = data_3['rating'].apply(handlerate)
```
